## **POSTS TO DATAFRAME**

In this section, we process Instagram posts stored in JSON format to transform them into a structured DataFrame. Each post's features are extracted and organized into corresponding columns, ensuring the data is both accessible and manageable for analysis.

The below code carefully handles nested structures and potential missing values, assigning default values where necessary. This transformation allows us to work with a clean and consistent dataset, simplifying further processing and enabling effective model training.

```python
import os
import json
import re
```

```python
dir_path = '/content/drive/My Drive/Colab Notebooks/extracted_20k_records/info/'
```

```python
post_data = []

# Iterate through all .info files in the directory
for filename in os.listdir(dir_path):
  if filename.endswith(".info"):
    file_path = os.path.join(dir_path, filename)

    # Open and read the .info file
    with open(file_path, 'r') as file:
      try:
        # Parse JSON content
        post = json.load(file)

        # Extract relevant fields
        post_data.append({
          'username': post.get('owner', {}).get('username', ''),  # Default to empty string if missing
          'post_id': post.get('id', ''),  # Default to empty string if missing
          'caption': (post.get('edge_media_to_caption', {})
                    .get('edges', [{}])[0]
                    .get('node', {})
                    .get('text', '') if post.get('edge_media_to_caption', {}).get('edges') else ''), # Handle deeply nested structure with defaults
          'likes': post.get('edge_media_preview_like', {}).get('count', 0),  # Default to 0 if missing
          'comments': post.get('edge_media_preview_comment', {}).get('count', 0),  # Default to 0 if missing
          'comments_parent': post.get('edge_media_to_parent_comment', {}).get('count', 0),  # Default to 0
          'timestamp': post.get('taken_at_timestamp', None),  # Default to None if missing
          'sponsorship': bool(post.get('edge_media_to_sponsor_user', {}).get('edges', [])),  # Boolean conversion
          'usertags': len(post.get('edge_media_to_tagged_user', {}).get('edges', [])),  # Count tagged users
          'location': post.get('location', ''),
          'accessibility_caption': post.get('accessibility_caption', ''),  # Default to empty string
          'is_verified': post.get('owner', {}).get('is_verified', False),  # Default to False
        })

        location = post.get('location')  # This will be None if location doesn't exist
        if location and isinstance(location.get('address_json'), str):
            try:
                location_data = json.loads(location.get('address_json', ''))
                post_data[-1]['location'] = location_data.get('country_code', '')
            except json.JSONDecodeError:
                post_data[-1]['location'] = ''  # Fallback if JSON is invalid
        else:
            post_data[-1]['location'] = ''  # If location is None or address_json is not a valid string

      except json.JSONDecodeError:
          print(f"Error decoding JSON in file: {filename}")
      except Exception as e:
          print(f"Unexpected error processing file {filename}: {e}")


# Create a DataFrame from the list of posts
df_posts = pd.DataFrame(post_data)
```

# **Feature Engineering for Model Readiness**

In this section, we process our data to extract and construct features that will enhance the model's ability to comprehend the dataset. Since many columns in the dataset contain strings or boolean values, we need to transform these into numerical representations or new derived features that capture meaningful information.

By creating additional columns with structured and model-friendly data, we enable the model to learn from patterns that might otherwise be obscured in raw formats. This step is crucial for improving predictive accuracy and ensuring the dataset is fully prepared for machine learning tasks.

```python
import pandas as pd
from datetime import datetime

data_path = "merge_data.csv"
df_merged = pd.read_csv(data_path)
```

This code transforms raw timestamps into meaningful categories by decoding Unix timestamps into readable dates and assigning them to time-of-day buckets **(Morning, Afternoon, Evening, Night)** and seasons **(Winter, Spring, Summer, Autumn)** based on the hour and month. These categorizations help analyze temporal patterns and trends in user activity or post engagement, and will be used as features for the predictive model to improve its accuracy and predictions.

```python
df_merged['decoded_timestamp'] = df_merged['timestamp'].apply(lambda x: datetime.utcfromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S') if pd.notnull(x) else None)

# Define a function to assign time-of-day buckets
def assign_time_bucket(decoded_timestamp):
    if pd.isnull(decoded_timestamp):
        return None
    hour = int(datetime.strptime(decoded_timestamp, '%Y-%m-%d %H:%M:%S').hour)
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

# Define a function to assign a season based on the month
def assign_season(decoded_timestamp):
    if pd.isnull(decoded_timestamp):
        return None
    month = int(datetime.strptime(decoded_timestamp, '%Y-%m-%d %H:%M:%S').month)
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'


# Create a new column for time-of-day buckets
df_merged['time_of_day'] = df_merged['decoded_timestamp'].apply(assign_time_bucket)
# Create a new column for seasons
df_merged['season'] = df_merged['decoded_timestamp'].apply(assign_season)
```

One Hot Encoding for the categorical columns

```python
df_merged = pd.get_dummies(df_merged, columns=["Category"], prefix="Category")
df_merged = pd.get_dummies(df_merged, columns=["time_of_day"], prefix="time_of_day")
df_merged = pd.get_dummies(df_merged, columns=["season"], prefix="season")
```

Turn boolean variables into numerics

```python
columns_to_convert = ['sponsorship', 'is_verified']
for col in columns_to_convert:
    df_merged[col] = df_merged[col].astype(bool).astype(int)
```

```python
from langdetect import detect
from googletrans import Translator
```

To ensure that the captions are uniformly processed, we detect their language and translate non-English captions to English. The following steps are performed:

1. Language Detection: The detect function identifies the language of the caption.
2. Translation: If the caption is not in English, it is translated to English using the Translator class.
3. Fallback Mechanism: If language detection or translation fails, the original caption is retained.

The transformed captions are then stored back in the caption column, ensuring that the dataset is ready for further processing and model training. This step helps the model by standardizing the textual input, improving its ability to learn from the captions.

```python
translator = Translator()

def detect_and_translate(caption):
    try:
        lang = detect(caption)
        if lang != 'en':
            caption = translator.translate(caption, src=lang, dest='en').text
        return caption
    except:
        return caption  # Return original if detection or translation fails

df_merged['caption'] = df_merged['caption'].fillna('').apply(detect_and_translate)
```

```python
import re
from nltk.corpus import stopwords
import string

import nltk
```

Cleaning Captions

```python
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def clean_caption(caption):
    # Remove URLs
    caption = re.sub(r'http\S+|www\S+|https\S+', '', caption, flags=re.MULTILINE)
    # Remove punctuation
    caption = caption.translate(str.maketrans('', '', string.punctuation))
    # Remove emojis
    caption = re.sub(r'[^\x00-\x7F]+', '', caption)
    # Remove stopwords
    caption = ' '.join([word for word in caption.split() if word.lower() not in stop_words])
    return caption

df_merged['caption_cleaned'] = df_merged['caption'].apply(clean_caption)
```

### **Extracting Sentiment Features**

Sentiment features provide the model with an understanding of the emotional tone and personal nature of the captions. This information can influence the number of likes a post receives, as positive or engaging captions are more likely to resonate with an audience.

```python
def extract_features(caption):
    features = {}
    sentiment = TextBlob(caption).sentiment
    features['sentiment_polarity'] = sentiment.polarity  # -1 (negative) to 1 (positive)
    features['sentiment_subjectivity'] = sentiment.subjectivity  # 0 (objective) to 1 (subjective)
    return features

caption_features = df_merged['caption_cleaned'].apply(extract_features).apply(pd.Series)
df_merged = pd.concat([df_merged, caption_features], axis=1)
```

The process involves cleaning and standardizing Instagram captions by removing unnecessary text (e.g., prefixes, placeholders, numbers) and filtering out common stopwords. **Lemmatization** is applied using `SpaCy` to reduce words to their base forms (e.g., "people" to "person") and eliminate redundant variations. This results in a simplified, meaningful  `cleaned_caption` column, ready for analysis or modeling with reduced noise and improved focus on relevant content.

```python
from nltk.tokenize import word_tokenize
import spacy

nltk.download('stopwords')
nltk.download('punkt')

# Cleaning the accessibility_caption column
df_merged['cleaned_caption'] = df_merged['accessibility_caption'].replace(
    "No photo description available.", ""
)  # Replace with blank

# Remove "Image may contain:" prefix
df_merged['cleaned_caption'] = df_merged['cleaned_caption'].str.replace(
    r"^Image may contain:\s*", "", regex=True
)  # Remove prefix

# Replace "text that says '...'" with "text"
df_merged['cleaned_caption'] = df_merged['cleaned_caption'].str.replace(r"text that says '.*?'", "text", regex=True)


# Custom stopwords list
custom_stopwords = set([
    "a", "an", "and", "the", "in", "on", "of", "is", "for", "to", "with",
    "by", "this", "that", "at", "as", "one", "more", "many", "or",",","?",","
])

# Function to clean captions
def clean_caption(caption):
    # Ensure the caption is a string
    if not isinstance(caption, str):
        return ''  # Return empty string if it's not a valid string

    # Remove numbers
    caption = re.sub(r'\b\d+\b', '', caption)

    # Tokenize and remove stopwords
    words = caption.split()
    filtered_words = [word for word in words if word.lower() not in custom_stopwords]

    # Rejoin the filtered words into a cleaned string
    return ' '.join(filtered_words)


# Apply the cleaning function to the accessibility_caption column
df_merged['cleaned_caption'] = df_merged['cleaned_caption'].apply(clean_caption)

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

# Function to lemmatize text
def lemmatize_caption_spacy(caption):
    doc = nlp(caption)
    return ' '.join([token.lemma_ for token in doc if token.is_alpha and not token.is_stop])

# Apply the function to the column
df_merged['cleaned_caption'] = df_merged['cleaned_caption'].apply(lemmatize_caption_spacy)
```

Identifies the 10 most frequent words from the captions and create new one-hot encoded columns for each word to represent their presence in individual captions.

First, it extracts the top 10 frequent words from the processed cleaned_caption text. Then, it defines a function to check if a specific word exists in a given caption. For each of the top 10 words, a new column is added to the DataFrame, where the value is 1 if the word is present in the caption and 0 otherwise.

This transforms the textual data into a structured format that can be easily used for further analysis or modeling.

```python
from collections import Counter

# Combine all captions into one string
all_captions = ' '.join(df_merged['cleaned_caption'])

# Tokenize into words
all_words = all_captions.split()

# Count the frequency of each word
word_counts = Counter(all_words)

# Get the 10 most frequent unique words along with their frequencies
most_frequent_unique_words = word_counts.most_common(10)

# Print the results
print("Top 10 most frequent unique words with frequencies:")
for word, freq in most_frequent_unique_words:
    print(f"{word}: {freq}")
```

```python
# List of the top 10 most frequent words
top_10_words = [word for word, count in most_frequent_unique_words]

# Function to perform one-hot encoding for each word
def one_hot_encode(caption, word):
    return 1 if word in caption.split() else 0

# Create one-hot encoded columns for each top word
for word in top_10_words:
    df_merged[f"contains_{word}"] = df_merged['cleaned_caption'].apply(lambda x: one_hot_encode(x, word))
```
